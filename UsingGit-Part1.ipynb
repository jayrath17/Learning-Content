# 3. Push one file from local to remote

We will create a text file that need to be pushed.

Now we have to push this file to remote.

before that we have to pull remote repo or clone it, we will use " git pull" 

The Git pull command is used to fetch and merge code changes from the remote repository to the local repository. 

Git pull is a combination of two commands, Git fetch followed by Git merge.


Before we pull, we need address of that remote repo.

As in the below image we will copy that https link of that repo

```python
# Now we use "git pull link" to pull repo to local machine and run this code in gitbash;
git pull https://github.com/jayrath17/testgit.git

#this command will pull all files(here README.md) to local machine.
```

```python
# Now we will run "git status" to check all untracked files.
git status

# And we can see in red-color that our 'test.txt' is untracked; 
```

```python
# so we will add it to staging area to commit it.
#for adding particular file; we use:
git add test.txt

#for adding all files:
git add .
```

```python
# Now we will commit it with some commit message:
git commit -m 'first commit'
```

```python
# Now we will push it to remote using command:
git push origin master

#here You can see error like this;
#it is telling that we have not given the address of remote repo where we have to push
```

```python
#we have to give origin address as below:
git remote add origin https://github.com/jayrath17/testgit.git
    
#then again run push command
git push origin master

#file will be pushed succesfuly
```

```python
#But here we can see in our gitbash, we are in 'master' branch and pushed to 'master' branch;
#So on remote, there is only one branch that is 'main';
#this 'push' will create new branch 'master' on remote repo and add 'test.txt' to that 'master' branch
```
